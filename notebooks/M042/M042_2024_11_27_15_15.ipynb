```python
%load_ext autoreload
%autoreload 2

import sys
sys.path.append("../..")

import pandas as pd
import numpy as np
import os

import matplotlib.pyplot as plt
import seaborn as sns
import pyaldata as pyal

import toolkit as tk
```

```python
data_dir = r"C:\data\raw\M042\M042_2024_11_27_15_15"
fname = os.path.join(data_dir, "M042_2024_11_27_15_15_pyaldata.mat")
```

```python
# load TrialData .mat file into a DataFrame
df = pyal.mat2dataframe(fname, shift_idx_fields=True)
```

## Preprocess

```python
df_ = tk.preprocess(df)
```

```python
df_.columns
```

```python
trial = 20
area = 'CP'
from mpl_toolkits.axes_grid1 import make_axes_locatable
with plt.style.context('seaborn-v0_8-bright'):
    sns.set_theme(context='talk', style='ticks', font='Arial')
    fig, ax = plt.subplots(1, 2, sharex='all', figsize=(12, 5))
    im1 = ax[0].imshow(df_[f"{area}_spikes"][trial].T, cmap='viridis')
    divider = make_axes_locatable(ax[0])
    cax = divider.append_axes("right", size="5%", pad=0.05)
    plt.colorbar(im1, cax=cax)
    
    im2 = ax[1].imshow(df_[f"{area}_rates"][trial].T, cmap='viridis')
    divider = make_axes_locatable(ax[1])
    cax = divider.append_axes("right", size="5%", pad=0.05)
    plt.colorbar(im2, cax=cax)
    for ax_ in ax:
        plt.title(f"{df_.loc[trial].values_Sol_direction}")
        ax_.axvline(x = df_.loc[trial].idx_sol_on, color='red', linewidth = '1')
        ax_.set_ylabel('Neurons')
        ax_.set_xlabel('Time (30 ms bins)')
        ax_.set_xticklabels([])
    plt.show()
```
